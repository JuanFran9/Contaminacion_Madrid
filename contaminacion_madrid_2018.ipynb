## 1.3 Importar .csv de todos los meses, filtrar por N02 y unir todos los datos de 2018 en una misma tabla 

```python
# Definir las rutas de los archivos para cada mes
file_paths = {
    'ene': 'Anio201810/ene_mo18.csv',
    'feb': 'Anio201810/feb_mo18.csv',
    'mar': 'Anio201810/mar_mo18.csv',
    'abr': 'Anio201810/abr_mo18.csv',
    'may': 'Anio201810/may_mo18.csv',
    'jun': 'Anio201810/jun_mo18.csv',
    'jul': 'Anio201810/jul_mo18.csv',
    'ago': 'Anio201810/ago_mo18.csv',
    'sep': 'Anio201810/sep_mo18.csv',
    'oct': 'Anio201810/oct_mo18.csv',
    'nov': 'Anio201810/nov_mo18.csv',
    'dic': 'Anio201810/dic_mo18.csv'
}

# Inicializar el DataFrame que contendrÃ¡ todos los datos
data_NO2 = pd.DataFrame()

# Cargar, filtrar y unir los datos de cada mes
for month, file_path in file_paths.items():
    # Cargar datos
    data_month = pd.read_csv(file_path, sep=';')
    # Filtrar por NO2
    data_month_NO2 = data_month[data_month['MAGNITUD'] == magnitud_NO2]
    # Unir los datos
    data_NO2 = pd.concat([data_NO2, data_month_NO2])

# Mostrar la informaciÃ³n del DataFrame consolidado
data_NO2
```

## 3.1 ARIMA 

```python
data_NO2_hourly_allyear
```

```python
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller


# Tomar solo la columna 'NO2_value' para el anÃ¡lisis de series temporales
ts_NO2 = data_NO2_hourly_allyear['NO2_value']

# Rellenar el valor NaN con el mÃ©todo 'ffill' que propaga el Ãºltimo valor vÃ¡lido hacia adelante
ts_NO2_filled = ts_NO2.fillna(method='ffill')

# Revisar nuevamente la cantidad de valores NaN para confirmar que el relleno fue exitoso
na_count_after_filling = ts_NO2_filled.isna().sum()


# FunciÃ³n para probar la estacionariedad
def test_stationarity(timeseries):
    # Determinar estadÃ­sticas mÃ³viles
    rolling_mean = timeseries.rolling(window=24*7).mean()  # ventana de 1 semana
    rolling_std = timeseries.rolling(window=24*7).std()

    # GrÃ¡fico de estadÃ­sticas mÃ³viles
    plt.figure(figsize=(14, 5))
    plt.plot(timeseries, color='blue',label='Original')
    plt.plot(rolling_mean, color='red', label='Rolling Mean')
    plt.plot(rolling_std, color='black', label = 'Rolling Std')
    plt.legend(loc='best')
    plt.title('Rolling Mean & Standard Deviation')
    plt.show()

    # Prueba Dickeyâ€“Fuller aumentada:
    print('Resultados de la Prueba de Dickey-Fuller:')
    dftest = adfuller(timeseries, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
    print(dfoutput)

# Si ya no hay valores NaN, repetir la prueba de estacionariedad
if na_count_after_filling == 0:
    test_stationarity(ts_NO2_filled)
else:
    print("AÃºn hay valores NaN despuÃ©s del relleno.")
```

#### El valor p de la preuba dickey fuller es extremadamente bajo (menos de 0.05) por lo que podemos confirmar su estacionariedad. Y por lo tanto, como no hace falta diferenciar, tendriamsos un parametro d=0 en el model ARIMA 

```python
ts_NO2_filled
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Plot ACF and PACF
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 8))

# Plot ACF
plot_acf(ts_NO2_filled, lags=50, ax=ax1)

# Plot PACF
plot_pacf(ts_NO2_filled, lags=50, ax=ax2, method='ywm')

plt.show()
```

#### hay una alta correlacion con las 12 y 24 horas anteriores como se puede observar en el grafico de autocorellacion, sorprendentemente los valores tardan bastante en llegar a estar dentro del humbral de confianza, lo que sugiere que podria haber menos estacionariedad de lo calculado anterirmente, por lo que habria que diferenciar.

## 3.2 RNN

### hubo problemas con los valores NaN, se podrian filtrar estos valores, remplazandolos con el valor anterior (ya que es serie temporal)

```python
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import SimpleRNN, Dense
from keras.optimizers import Adam
import numpy as np

# Definir el nÃºmero de pasos atrÃ¡s (lag)
n_lag = 24

NO2_values = data_NO2_hourly_allyear['NO2_value'].values.reshape(-1, 1)  # Convertir la serie en una matriz 2D para el escalador


# Escalar los valores de NO2 sin problemas
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(NO2_values)

# FunciÃ³n para crear secuencias de datos
def create_dataset(data, n_lag):
    dataX, dataY = [], []
    for i in range(len(data) - n_lag - 1):
        a = data[i:(i + n_lag), 0]
        dataX.append(a)
        dataY.append(data[i + n_lag, 0])
    return np.array(dataX), np.array(dataY)

# Crear datos de entrenamiento y prueba
train_size = int(len(scaled_data) * 0.8)
test_size = len(scaled_data) - train_size
train, test = scaled_data[0:train_size,:], scaled_data[train_size:len(scaled_data),:]
trainX, trainY = create_dataset(train, n_lag)
testX, testY = create_dataset(test, n_lag)

# Reformar input para ser [muestras, pasos de tiempo, caracterÃ­sticas]
trainX = np.reshape(trainX, (trainX.shape[0], n_lag, 1))
testX = np.reshape(testX, (testX.shape[0], n_lag, 1))

# Crear y ajustar la Red SimpleRNN
model = Sequential()
model.add(SimpleRNN(units=50, activation='tanh', input_shape=(n_lag, 1)))
model.add(Dense(1, activation='linear'))
model.compile(loss='mean_squared_error', optimizer=Adam(lr=0.01))
model.fit(trainX, trainY, epochs=50, batch_size=72, validation_data=(testX, testY), verbose=1)

# Realizar predicciones
trainPredict = model.predict(trainX)
testPredict = model.predict(testX)
```
